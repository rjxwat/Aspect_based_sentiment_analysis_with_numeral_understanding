```python
import pandas as pd
import nltk
import re
import string
from nltk.corpus import stopwords
from nltk.stem import WordNetLemmatizer

lemmatizer = WordNetLemmatizer()
```

```python
def preprocess_text(text):
    # Lowercase the text
    text = text.lower()
    
    text = re.sub(r'http\S+|www\S+|https\S+', '', text, flags=re.MULTILINE)
    
    text = text.translate(str.maketrans('', '', string.punctuation))
    

    
    text = re.sub(r'\s+', ' ', text).strip()
    
    tokens = nltk.word_tokenize(text)
    
    stop_words = set(stopwords.words('english'))
    tokens = [word for word in tokens if word not in stop_words]
    
    tokens = [lemmatizer.lemmatize(word) for word in tokens]
    
    
    return ' '.join(tokens)
```

```python

# Apply preprocessing

for i in range(len(data['news'])):

    data['cleaned_text'][i]= preprocess_text(data['news'][i])

# Display the cleaned DataFrame
print(data[['news', 'cleaned_text']])
```

```python
import pandas as pd
import torch
from transformers import BertTokenizer, BertModel

# Load your dataset
# Assuming your dataset is a CSV file with a column 'text'
t = data['cleaned_text']  # Select the relevant column

# Initialize the BERT tokenizer and model
tokenizer = BertTokenizer.from_pretrained('bert-base-uncased')
model = BertModel.from_pretrained('bert-base-uncased')
model.eval()  # Set the model to evaluation mode
```

```python


# Function to get embeddings
def get_embeddings(texts):
    embeddings = []
    for text in texts:
        # Tokenize and create tensor
        inputs = tokenizer(text, truncation=True, padding='max_length', max_length=512, return_tensors='pt')
        
        # Get embeddings
        with torch.no_grad():
            outputs = model(**inputs)
        
        # Get the last hidden state (you can also use pooled output)
        last_hidden_state = outputs.last_hidden_state  # Shape: (batch_size, sequence_length, hidden_size)
        
        # Average the token embeddings to get a single vector (you can use other methods as well)
        text_embedding = last_hidden_state.mean(dim=1).squeeze().cpu().numpy()
        embeddings.append(text_embedding)
    
    return embeddings


```

```python
# Get embeddings for all text entries
data['embeddings'] = get_embeddings(data['cleaned_text'].tolist())

# Save the embeddings to a CSV file
# data.to_csv('text_embeddings.csv', index=False)

# print("Embeddings saved to 'text_embeddings.csv'")
```
